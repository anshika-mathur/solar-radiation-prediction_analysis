# **Linear Regression Model Implementation without outliers->**

```python
cleaned_data.info()
```

```python
cleaned_data.columns
```

```python
y = cleaned_data['Surface Albedo']
```

```python
X = cleaned_data[['Year', 'Month', 'Day', 'Temperature', 'Clearsky DHI', 'Clearsky DNI',
       'Clearsky GHI', 'Dew Point', 'DHI', 'DNI', 'GHI', 'Relative Humidity',
       'Solar Zenith Angle', 'Pressure', 'Wind Speed']]
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test ,y_train,y_test= train_test_split(X,y,train_size=0.3,random_state=0)
```

```python
X_train.shape , X_test.shape ,y_train.shape ,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
c=model.intercept_
c
```

```python
m= model.coef_
m
```

```python
y_pred_train = m * X_train+c
y_pred_train
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
y_pred.shape
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
mae= mean_squared_error(y_test,y_pred)
mse= mean_absolute_error(y_test,y_pred)
mae
```
